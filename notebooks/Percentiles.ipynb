# Plot percentiles Peso

```python
weight.query('Sex==1')[['P3', 'P5', 'P10', 'P25', 'P50', 'P75', 'P90', 'P95', 'P97']].plot(figsize=(15,10))
plt.title("Percentiles Peso Niños")
plt.xlabel("Edad")
plt.ylabel("Peso")
plt.show()
```

```python
weight.query('Sex==2')[['P3', 'P5', 'P10', 'P25', 'P50', 'P75', 'P90', 'P95', 'P97']].plot(figsize=(15,10))
plt.title("Percentiles Peso Niñas")
plt.xlabel("Edad")
plt.ylabel("Peso")
plt.show()
```
